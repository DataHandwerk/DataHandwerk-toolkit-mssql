## Import JSON documents into SQL Server

### Import a JSON document into a single column

```sql
SELECT BulkColumn
 FROM OPENROWSET (BULK 'D:\Users\ggoer\OneDrive\dbeaver\dhw\General\.dbeaver\data-sources.json', SINGLE_CLOB) as j;
```

### Parse JSON documents into rows and columns

```sql
SELECT value
 FROM OPENROWSET (BULK 'D:\Users\ggoer\OneDrive\dbeaver\dhw\General\.dbeaver\data-sources.json', SINGLE_CLOB) as j
 CROSS APPLY OPENJSON(BulkColumn)
```

```sql
SELECT *
 FROM OPENROWSET (BULK 'D:\Users\ggoer\OneDrive\dbeaver\dhw\General\.dbeaver\data-sources.json', SINGLE_CLOB) as j
 CROSS APPLY OPENJSON(BulkColumn)
```

```sql
TRUNCATE TABLE repo.dbeaver_DataSources

INSERT INTO repo.dbeaver_DataSources (BulkColumn)
SELECT BulkColumn
FROM OPENROWSET(BULK 'D:\Users\ggoer\OneDrive\dbeaver\dhw\General\.dbeaver\data-sources.json', SINGLE_CLOB) AS j;
```

```sql
SELECT 
--
j1.*
 , j2.*
FROM [repo].[dbeaver_DataSources] j1
CROSS APPLY OPENJSON(BulkColumn) j2
WHERE [is_json] = 1
```

```sql
SELECT 
--
j1.*
 , j2.*
 , j3.*
FROM [repo].[dbeaver_DataSources] j1
CROSS APPLY OPENJSON(BulkColumn) j2
CROSS APPLY OPENJSON(j2.value) j3

WHERE [is_json] = 1
```

```sql
SELECT
 --
 j1.*
 , j2.*
 , j3.[key] AS connection_key
 , j3.value AS connection_value
 , J4.*
FROM [repo].[dbeaver_DataSources] j1
CROSS APPLY OPENJSON(BulkColumn) WITH (connections NVARCHAR(MAX) N'$.connections' AS JSON) j2
CROSS APPLY OPENJSON(j2.connections) j3
CROSS APPLY OPENJSON(j3.value) WITH (
  connection_provider NVARCHAR(100) N'$.provider'
  , connection_driver NVARCHAR(100) N'$.driver'
  , conection_name NVARCHAR(100) N'$.name'
  ) j4
```

```sql
SELECT
 --
 vm.*
FROM (
 SELECT 1 AS 'virtual-models'
 ) vm
FOR JSON AUTO
 , WITHOUT_ARRAY_WRAPPER
```

```sql
SELECT
 --
 vm.*
FROM (
 SELECT
  (
   SELECT
    --
    instance.*
   FROM (
    SELECT 1 AS 'blablabla'
    ) instance
   FOR JSON AUTO
    , WITHOUT_ARRAY_WRAPPER
   ) AS 'virtual-models.DbeaverIdDerServerInstance'
 ) vm
FOR JSON PATH
 , WITHOUT_ARRAY_WRAPPER
```

Das Problem mit dem JSON PATH ist: man kann zwar einen Pfad angeben, muss den aber als Literal definieren  
Also müsste man erst erst mal das SELECT Statement generieren, dass dann die Namen der Instanzen, Datenbank(en) und Schemas und entitäten enthält

```sql
SELECT 'ccc' AS 'virtual-models.microsoft-177b4c65f69-4b8d1d14f182fa5d.dhw_self.repo.:ForeignKey_virtual.constraints'
 , (
  SELECT entity = [referenced_RepoObject_fullname]
   , [constraint] = [ForeignKey_name]
   , attributes = [referencing_IndexPatternColumnName]
  FROM [repo].[ForeignKey_Index_guid_union] AS fk
  WHERE fk.[referencing_SysObject_name] = 'ForeignKey_virtual'
   AND fk.[referencing_SysObject_schema_name] = 'repo'
  FOR json auto
  ) AS 'virtual-models.microsoft-177b4c65f69-4b8d1d14f182fa5d.dhw_self.repo.:ForeignKey_virtual.foreign-keys'
FOR JSON PATH
 , WITHOUT_ARRAY_WRAPPER
```
